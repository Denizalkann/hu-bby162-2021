```python
#Kütüphane Otomasyonu
#Veri Tabanı: /content/drive/MyDrive/Colab Notebooks/Final.txt

dosya = "/content/drive/MyDrive/Colab Notebooks/Final.txt"
print("Kütüphane Kataloglama Sistemine Hoşgeldiniz")

def eserListele():
  f = open(dosya, 'r')
  for line in f.readlines():
    print(line)
  f.close()

def eserKaydet():
  sisKulAdi = "admin"
  sisKulSifre = "admin"
  girKulAdi = input("Kullanıcı Adı: ")
  girKulSifre = input("Şifre: ")
  if girKulAdi == sisKulAdi and girKulSifre == sisKulSifre:
    print("Yeni bir kayıt girin.\n")
  eserAdi = input("Eser Adını Girin.")
  eserYazar = input("Eserin Yazarını Girin.")
  eserYayınevi = input("Eserin Yayınevi Bilgisini Girin.")
  eserBasımtarihi = input("Eserin Basım Tarihini Girin (Yıl)")
  eserIsbn = input("Eserin ISBN Bilgilerini Girin.")
  f = open(dosya, 'a' )
  f.write(eserAdi + " , " + eserYazar + " , " + eserYayınevi + " , " + eserBasımtarihi + " , " + eserIsbn + '\n')
  print("Kaydı Başarıyla Eklediniz")
  f.close()
  
islemSecme = ["listele", "ekle", "bitir"]
while islemSecme:
  girilen_islem = input("Katalogdaki eser listesini görüntülemek için 'listele'\n"
      "Yeni eser kaydı eklemek için 'ekle' görevini yazın.\n"     
      "Sistemden çıkmak için 'bitir' görevini yazın.\n"
      "Yapmak istediğiniz işlem için ilgili görevi yazın: ")
  
  if girilen_islem == islemSecme[0]:
    print("Kütüphane kataloğuna hoşgeldiniz..")
    eserListele()
  elif girilen_islem == islemSecme[1]:
    print("Kataloğa yeni bir kayıt girin.")
    eserKaydet()
  elif girilen_islem == islemSecme[2]:
    break
```
